```python
# EXPLORATORY DATA ANALYSIS
ax = sbn.countplot(x='Gender',data=df)
for bar in ax.containers:
  ax.bar_label(bar)
```

```python
date=df.groupby(['Gender'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
```

```python
bx=sbn.barplot(x='Gender',y='Amount',data = date)
```

```python
second_analysis=df.groupby(['Age Group'],as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=True)
```

```python
second_analysis
```

```python
sbn.barplot(x='Age Group', y='Amount',data=second_analysis)
```
